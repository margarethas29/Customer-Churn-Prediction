```python
#Membagi fitur dataset menjadi categorical dan numerical kemudian mencari hubungannya dengan target
categorical = ['area_code', 'international_plan', 'voice_mail_plan', 'churn', 'state']
numerical = ['account_length', 'number_vmail_messages', 'total_day_minutes', 'total_day_calls', 'total_day_charge', 'total_eve_minutes', 'total_eve_calls', 'total_eve_charge', 'total_night_minutes', 'total_night_calls', 'total_night_charge', 'total_intl_minutes', 'total_intl_calls', 'total_intl_charge', 'number_customer_service_calls']
```

```python
#Target
def plot_pie_chart(data):
    """
    Fungsi untuk membuat pie chart dari kolom churn dalam data pelanggan.

    Parameter:
    - data: DataFrame Pandas yang berisi kolom churn

    Output:
    - Menampilkan pie chart dari kolom churn
    """
    # Hitung jumlah pelanggan yang churn dan tidak churn
    data['churn'] = data['churn'].map({'yes': 1, 'no': 0})

    # Hitung jumlah pelanggan yang churn dan tidak churn
    churn_counts = data['churn'].value_counts()

    # Buat pie chart
    plt.figure(figsize=(6, 4))
    plt.pie(churn_counts, labels=churn_counts.index.map({1: 'Yes', 0: 'No'}), autopct='%1.1f%%', startangle=140)
    plt.title('Pie Chart of Churn')
    plt.axis('equal')
    plt.show()
```

```python
plot_pie_chart(data_train)
"""Dapat dilihat bahwa Churn "NO" pada chart sebesar 85.9% di mana jika kita memprediksi akan cenderung no churn atau pelanggan tidak berpindah."""
```

#Pre-Processing Data

####Encoding Data

```python
# Untuk Label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_train['international_plan'] = le.fit_transform(data_train['international_plan'])
data_train['voice_mail_plan'] = le.fit_transform(data_train['voice_mail_plan'])
data_train['churn'] = le.fit_transform(data_train['churn'])
```

```python
# Hitung frekuensi kemunculan setiap nilai dalam kolom 'state'
state_counts = data_train['state'].value_counts()

# Hitung panjang DataFrame
data_length = len(data_train)

# Buat dictionary untuk mapping nilai state ke frekuensi kemunculannya, dibagi dengan panjang DataFrame
state_freq_map = (state_counts / data_length).to_dict()

# Ganti nilai dalam kolom 'state' dengan frekuensinya
data_train['state_freq'] = data_train['state'].map(state_freq_map)
```

```python
# Lakukan one-hot encoding pada kolom 'area_code'
area_code_encoded = pd.get_dummies(data_train['area_code'], prefix='is')

# Gabungkan hasil encoding dengan DataFrame asli
data_train_encoded = pd.concat([data_train, area_code_encoded], axis=1)
```

```python
data_train_encoded.head()
```

####Normalisasi Data

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
```

```python
#drop kolom state, area_code pada data train

data_train_encoded.drop(["state","area_code"], inplace=True,axis=1)
```

```python
data_train_encoded.head()
```

```python
data_normalized = minmax_scaler.fit_transform(data_train_encoded)
```

```python
df_train = pd.DataFrame(data_normalized, columns=data_train_encoded.columns)
```

```python
df_train.head()
```

####Splitting Data

```python
from sklearn.model_selection import train_test_split

x = df_train.drop(['churn'], axis=1)
y = df_train['churn']

#Data dibagi menjadi 80:20
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=123, test_size=0.2)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

#Pemodelan Data

###K-NN

```python
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

###Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

###Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
```

#Pelatihan Model

```python
from sklearn.metrics import classification_report
```

###K-NN

```python
print("Accuracy: %.2f%%" % (result1*100.0))
```

```python
print(classification_report(y_test, knn.predict(x_test)))
```

###Decision Tree

```python
result2 = dtc.score(x_test, y_test)
```

```python
print("Accuracy: %.2f%%" % (result2*100.0))
```

```python
print(classification_report(y_test, dtc.predict(x_test)))
```

###Random Forest

```python
result3 = rfc.score(x_test, y_test)
```

```python
print("Accuracy: %.2f%%" % (result3*100.0))
```
